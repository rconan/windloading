```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("dark_background")
```

```python
df = pd.read_parquet("/fsx/CASES/zen30az000_OS7/windloading.parquet")
```

```python
df.head()
```

```python
len(df)
```

```python
m1_rbm = np.swapaxes(np.vstack(df["M1RigidBodyMotions"]).reshape(-1,7,6),2,1)
m2_rbm = np.swapaxes(np.vstack(df["M2RigidBodyMotions"]).reshape(-1,7,6),2,1)
m1_bm = np.swapaxes(np.vstack(df["M1BendingModes"]).reshape(-1,7,162),2,1)
```

```python
sid = 1
t = np.arange(len(df))*1e-3
fig, axs = plt.subplots(nrows=3, sharex=True, figsize=(10,15))
ax = axs[0]
ax.plot(t,m1_rbm[...,sid-1]);
ax.grid()
ax = axs[1]
ax.plot(t,m2_rbm[...,sid-1]);
ax.grid()
ax = axs[2]
ax.plot(t,m1_bm[...,sid-1]);
ax.grid()
```
